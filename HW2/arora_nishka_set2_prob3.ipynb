```python
import numpy as np
from matplotlib import pyplot as plt
from numpy import genfromtxt
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
%matplotlib inline
```

## Load data

```python
# Upload problem3data.txt to a 'data' folder in Google Colab, using the "Files"
# tab on the left.
# (You can also use urllib.request.urlretrieve, as in set 1)
train_file = 'https://raw.githubusercontent.com/charlesincharge/Caltech-CS155-2022/main/sets/set2/data/problem3data.txt'
train_data = genfromtxt(train_file, delimiter='\t')

y_train = train_data[:, 9]
X_train = train_data[:, :9]
```

```python
lasso_coefs = []
ridge_coefs = []

for alp in np.linspace(0.01, 3, 30):
  lasso_clf = Lasso(alpha=alp)
  lasso_clf.fit(X_train, y_train)
  lasso_coefs.append(lasso_clf.coef_)

for alp in range(1, 10001):
  ridge_clf = Ridge(alpha=alp)
  ridge_clf.fit(X_train, y_train)
  ridge_coefs.append(ridge_clf.coef_)
```

```python
plt.figure()
plt.plot(np.linspace(0.01, 3, 30), lasso_coefs)
plt.title('w vs alpha')
plt.xlabel('alpha')
plt.ylabel('w')
plt.legend(['w1', 'w2', 'w3', 'w4', 'w5', 'w6', 'w7', 'w8', 'w9'], loc='right')
plt.savefig('Lasso values of w')

plt.figure()
plt.plot(np.linspace(1, 10000, 10000), ridge_coefs)
plt.title('w vs alpha')
plt.xlabel('alpha')
plt.ylabel('w')
plt.legend(['w1', 'w2', 'w3', 'w4', 'w5', 'w6', 'w7', 'w8', 'w9'], loc='right')
plt.savefig('Ridge values of w')
```
